## Table Entity Relationship Diagram

The table diagram for the relationships between CSV's
was designed so that, when all the information was put together,
the relationships between each CSV is clearly outlined and 
the final product is easy to understand.

## QuickDBD - http://www.quickdatabasediagrams.com

QuickDBD was used as the visualization tool for the ERD. The resulting
graph is fairly intuitive and represents the relationships between
Primary Keys, Foreign Keys, and Unique Keys. The titles of the boxes are
the .csv file names, and the lists under them are headers within the files.



## Tables

This section outlines relationships chosen between headers on
CSV's and other CSV's. CSV titles will be ennumerated and then
a list of headers with their particular relationship to other tables
will be listed. Three key types were used:

* Primary keys (PK) are generally chosen when they define the CSV that they're
a part of, such as department number inside the department CSV.


* Foreign keys (FK) are chosen when there are columns with the same information
in other CSVs but they aren't unique to the CSV that is being worked with,
such as employee number under the salary table, versus employee number under
the employee table.


* Unique keys (UNIQUE) are present only within that CSV and provide information
not found anywhere else, such as salary under the salaries table, or
titles for employees under the titles table.

#### 1. department.csv

* dept_no PK
* dept_name UNIQUE

#### 2. dept_emp.csv 

* emp_no FK
* dept_no FK
* from_date DATE FK
* to_date DATE FK

#### 3. employees.csv

* emp_no PK
* birth_date UNIQUE
* first_name UNIQUE
* last_name UNIQUE
* gender UNIQUE
* hire_date UNIQUE

#### 4. salaries.csv

* emp_no FK
* salary UNIQUE
* from_date FK
* to_date FK

#### 5. titles.csv

* emp_no FK
* title UNIQUE
* from_date FK
* to_date FK

#### 6. dept_manager.csv

* dept_no FK
* emp_no FK
* from_date FK
* to_date FK

## Importation from PostgreSQL and Data Engineering Information

The databases were joined using SQL commands and the schema illustrated above, and by using CREATE VIEW to store the queries. Four tables were created, 2 of them are titles and salaries with no duplicates for the same employee id, and 2 are resulting databases from joining all of the .csv files, one with duplicates and one without. 



Using SQL Alchemy, the database was imported from postgres. For the salary graph, the table without duplicate entries for employees and only the most up to date position and salary was used.

```python
import os
import matplotlib.pyplot as plt
import sqlalchemy as db
import pandas as pd
password = os.environ.get('DB_PASS')
engine = db.create_engine("postgresql+psycopg2://postgres:"+password+"@/PewlettHackard")
connection = engine.connect()
```

 After the engine was created, the pd.read_sql_table() command was used in order to import the table necessary for analysis. Salary column was converted to float64 from object as it was necessary for the groupby function. NaT under to_date means that the employee is currently working for the company as the year 9999 doesn't translate from SQL even though it's represented as such when .dtypes is used

```python
full_database = pd.read_sql_table('full_database', engine)
full_database.head()
```

```python
full_database['salary'] = full_database['salary'].astype('float')
full_database.dtypes
```

Below, the data was grouped by title and then a count of employees and an average of salary was taken to show company size and payscale.

```python
salary_mean = full_database.groupby(['title'])['salary'].mean()
employee_title_mean = full_database.groupby(['title'])['emp_no'].count()
```

```python
employee_title_mean_df = pd.DataFrame(employee_title_mean)
employee_title_mean_df = employee_title_mean_df.reset_index()
employee_title_mean_df.columns = ['Title','Employee Count']
employee_title_mean_df
```

```python
salary_mean_df = pd.DataFrame(salary_mean)
salary_mean_df = salary_mean_df.reset_index()
salary_mean_df['salary'] = salary_mean_df['salary'].round(1)
salary_mean_df.columns = ['Title','Salary']
salary_mean_df
```

### Finally, the information was graphed to compare Salary vs. Title and Employee Count vs. Title.

What tips off that this data set is fake is the rather flat distribution of salary given different positions held. You'd think that with an increase in responsibility there would be an increase in compensation, but it seems as though that management makes the same or less than the people they're in charge of. Also, the senior staff seems to surpass every other category which is also rather strange.

```python
salaryvstitle = salary_mean_df.plot(kind='bar',
                                    x='Title',
                                    y='Salary', 
                                    color = 'peru',
                                    edgecolor='black', 
                                    alpha=.7, 
                                    legend = False)
salaryvstitle.set_xlabel('Title')
salaryvstitle.set_ylabel('Salary')
salaryvstitle.set_facecolor('azure')
salaryvstitle.set_title ('Salary vs. Title')

salaryvstitle = salaryvstitle.get_figure()

salaryvstitle.savefig('Documents\GitHub\PewlettHackard\jupyter notebook graphs\salary-vs-title.png',bbox_inches="tight")
```

```python
employee_vs_title = employee_title_mean_df.plot(kind='bar',x='Title',y='Employee Count', 
                                                color = 'peru',
                                                edgecolor='black', 
                                                alpha=.7, 
                                                legend = False)
employee_vs_title.set_xlabel('Title')
employee_vs_title.set_ylabel('Employee Count')
employee_vs_title.set_facecolor('azure')
employee_vs_title.set_title ('Employee Count vs. Title')

employee_vs_title = employee_vs_title.get_figure()

employee_vs_title.savefig('Documents\GitHub\PewlettHackard\jupyter notebook graphs\employee-count.png', bbox_inches="tight")
```
